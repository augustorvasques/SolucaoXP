```python
def organizar_ordens(nome_arquivo):
    df = pd.DataFrame(columns= ['Código', 'Natureza', 'Quantidade', 'Ativo', 'Preço', 'Garantia Disp'])
    arquivo = open(nome_arquivo, "r")
    novas_linhas = []

    # Ler linha por linha
    ordem = arquivo.readline()
    while ordem != "":
        # Dividir a ordem em seus dados
        valores = ordem.strip().split(';')
        # Criar um dicionário para a ordem
        novo_dado = {
            'Código': int(valores[0]),
            'Natureza': valores[1],
            'Quantidade': int(valores[2]),
            'Ativo': valores[3],
            'Preço': float(valores[4]),
            'Garantia Disp': float(valores[5])
        }

        # Adicionar o dicionário à lista de novas linhas
        novas_linhas.append(novo_dado)
        # Ler a próxima linha
        ordem = arquivo.readline()

    # Fechar o arquivo
    arquivo.close()

    # Converter a lista de dicionários em um DataFrame e concatenar com o DataFrame original
    df_novas_linhas = pd.DataFrame(novas_linhas)
    df = pd.concat([df, df_novas_linhas], ignore_index=True)
    
    # Cálculo/Inclusão das novas colunas necessárias
    desagio = [0.05 if x == 'PETR4' else 0.1 for x in df['Ativo']]
    df['Deságio'] = desagio
    limite_fin = [100000 if x == 'PETR4' else 80000 for x in df['Ativo']]
    df['Limite Boleta'] = limite_fin
    df['Garantia Exigida'] = df['Quantidade']* df['Preço']*df['Deságio']
    df['Resultado Garantia'] = df.apply(lambda row: 'Rejeitado' if row['Garantia Exigida'] > row['Garantia Disp'] else 'Aprovado', axis=1)
    df['Volume Boleta'] = df['Quantidade'] * df['Preço']
    df['Resultado Boleta'] = df.apply(lambda row: 'Rejeitado' if row['Volume Boleta'] > row['Limite Boleta'] else 'Aprovado', axis=1)
    
    
    
    return df
```

```python
# Teste primeira func
organizar_ordens('Ex_Clientes_Ordens.txt')
```

```python
def analisar_ordens(nome_arquivo):
    df = organizar_ordens(nome_arquivo)
    
    qtde_rejeitadas = 0
    qtde_enviadas = 0
    ordens_rejeitadas = []
    ordens_aprovadas =[]
    i = 1
    
    # Checagem das ordens
    for index, row in df.iterrows():
        # Checagem garantia
        if row['Resultado Garantia'] == 'Rejeitado':
            print(f'Ordem da linha {i} do cliente {row["Código"]} foi \033[31mrejeitada\033[m, pois garantia exigida de {row["Garantia Exigida"]} ultrapassou a garantia disponível de {row["Garantia Disp"]}!\n')
            ordens_rejeitadas.append(i)
            i += 1
        # Checagem volume boleta    
        elif row['Resultado Boleta'] == 'Rejeitado':
            print(f'Ordem da linha {i} do cliente {row["Código"]} foi \033[31mrejeitada\033[m, pois o volume financeiro da boleta de {row["Volume Boleta"]} ultrapassou o limite máximo definido de {row["Limite Boleta"]}\n')
            ordens_rejeitadas.append(i)
            i += 1
        # Caso chegar aqui, ordem aprovada!    
        else:
            print(f'\033[32mOrdem da linha {i} do cliente {row["Código"]} foi enviada com sucesso!\033[m\n')
            ordens_aprovadas.append(i)
            i += 1
    
    # Dict com ordens aprovadas e reprovadas para facilitar usos futuros e localização no df original
    resultado_ordens = {'Aprovadas': ordens_aprovadas, 'Reprovadas': ordens_rejeitadas}
    
    return resultado_ordens
```
